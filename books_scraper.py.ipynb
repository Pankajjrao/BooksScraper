```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python


def rating_func(article_tag):
    rating = tag.find("p")['class'][1]
    return rating
    
def name_func(article_tag):
    name = tag.find('h3').a['title']
    return name

def cost_func(article_tag):
    cost = tag.find("p",class_='price_color').text[2:]
    return cost

def stock_func(article_tag):
    available = tag.find("p",class_= "instock availability")
    availability = True if available else False
    return availability

all_books = []
page_number = 1

while True:
    url = f"https://books.toscrape.com/catalogue/page-{page_number}.html"
    source = requests.get(url, headers=headers,).text
    
    soup = BeautifulSoup(source, "html.parser")

    all_article_tags = soup.find_all("article",class_="product_pod")
    
    print(page_number)

    for tag in all_article_tags:

        #name
        name = name_func(tag)
        
        # ratings
        rating = rating_func(tag)
    
        #price
        cost = cost_func(tag)
        

        # instock
        availability = stock_func(tag)

        data = {"Name" : name,"Price" : cost, "Rating" : rating, "Instock" : availability}
        
        all_books.append(data)
    next_page = soup.find('li',class_='next')
    if next_page:
        page_number += 1
    else:
        break
    
    
    
    
    
```

```python
df = pd.DataFrame.from_dict(all_books)
df.to_csv("Books.csv",index = False)
```
